# Now we can try to check out some origin of those outliers 

```python
day1=Clusters[Clusters['NBRE DE JOUR']==1]
day1_greater24h=day1[day1['NBRE HEURES DE FONCTIONNEMENT DG1']>24]
day1_greater24h.head(10)
```

The above table show that there are some running time of the generator (NBRE HEURES DE FONCTIONNEMENT DG1) are above 24 hours in one days.

# Here we assume that a generator can't work more than 24hours in one day

```python
import time
start_time = time.time()
Clusters["Normal_running_time"]=0  #we create and initialise the new column with 0
import numpy as np
for x in np.arange(Clusters.shape[0]):
    if Clusters['NBRE DE JOUR'][x]!=0:
        Clusters["Normal_running_time"][x]=Clusters['NBRE HEURES DE FONCTIONNEMENT DG1'][x]/Clusters['NBRE DE JOUR'][x]
        if Clusters["Normal_running_time"][x]>24:         #check if running time normal
            Clusters["Normal_running_time"][x]=24*Clusters['NBRE DE JOUR'][x] #we assign the normal running time
        else:
            Clusters["Normal_running_time"][x]=Clusters['NBRE HEURES DE FONCTIONNEMENT DG1'][x]            
print("--This code take %s minute(s) to run--" % ((time.time() - start_time)/60))
      
```

```python
X=data_num
Y=data_all_var["Normal_cons_HIS"]
validation_size = 0.30
seed = 7
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=validation_size, random_state=seed)
# Test options and evaluation metric
num_folds = 5

#Spot-Check Algorithms
models = []
models.append(('LR',LinearRegression()))
models.append(('RF',RandomForestRegressor()))
models.append(('Gbt',GradientBoostingRegressor()))
models.append(('KNN',KNeighborsRegressor()))
models.append(('ETR',ExtraTreesRegressor()))
models.append(('DTR',DecisionTreeRegressor()))
models.append(('AdR',AdaBoostRegressor()))
models.append(('MLP',MLPRegressor()))
clf=RandomForestRegressor()
pred=clf.fit(X_test,Y_test)
predicted=clf.predict(X_test)
results=[]
names=[]
for name, model in models:
    kfold = KFold(n_splits=num_folds, random_state=seed)
    cv_results = cross_val_score(model, X_train, Y_train, cv=kfold)
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" %(name, cv_results.mean(), cv_results.std())
    print(msg)
```
